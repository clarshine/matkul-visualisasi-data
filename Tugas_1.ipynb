### **Import Library yang dibutuhkan**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### **Import haberman.csv dan assign dataset ke variable cancer_sur**

```python
cancer_sur = pd.read_csv('haberman.csv', header=None, names=["age", "operation_year", "axil_nodes_det", "surv_status"])
```

```python
cancer_sur.head()
```

### **Deskripsi nilai statistika berupa mean, min, max, dll**

```python
print(cancer_sur.describe())
```

### **Cek jumlah baris dan kolom pada dataset**

```python
cancer_sur.shape
```

### **Cek jumlah pasien yang selamat dan yang meninggal**

```python
cancer_sur["surv_status"].value_counts()
```

# **Univariate Analysis**

### **Plot Distribusi surv_status terhadap age**

```python
sns.FacetGrid(cancer_sur, hue="surv_status", height=5).map(sns.distplot, "age").add_legend()
```

### **Plot distribusi surv_status terhadap operation_year**

```python
sns.FacetGrid(cancer_sur, hue="surv_status", height=5).map(sns.distplot, "operation_year").add_legend()
```

### **Probability Density Function (PDF) dan Cumulative Distribution Function (CDF) dari umur pasien**

```python
counts, bin_edges = np.histogram(cancer_sur['age'],bins=10,density=True)
plt.xlabel('age')
pdf = counts/(sum(counts))
print("pdf=", pdf);
print("bin_edges", bin_edges);
cdf = np.cumsum(pdf)
print("cdf=", cdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
```

# **Bivariate Analysis**

### **BoxPlot yang menggambarkan hubungan antara value surv_status terhadap age**

```python
sns.boxplot(x="surv_status", y="age", data=cancer_sur)
plt.show()
```

### **BoxPlot yang menggambarkan hubungan surv_status terhadap operation_year**

```python
sns.boxplot(x="surv_status", y="operation_year", data=cancer_sur)
plt.show()
```

### **ViolinPlot untuk visualisasi data terhadap nilai PDF**

```python
#ViolinPlot
sns.violinplot(x="surv_status", y="operation_year", data=cancer_sur)
plt.show()
```

# **Multivariate Analysis**

### **Pair Plot**

```python
sns.pairplot(cancer_sur, hue="surv_status", height=3)
plt.show()
```
